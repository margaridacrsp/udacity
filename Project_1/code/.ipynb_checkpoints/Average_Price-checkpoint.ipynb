```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import datetime
import seaborn as sns
```

```python
#load the calendar availability and prices data
calendar = pd.read_csv('calendar.csv')
calendar.shape
```

```python
#load the listings data
listings = pd.read_csv('listings.csv')
listings.shape
```

```python
#view the data in calendar dataframe
calendar['date'] = pd.to_datetime(calendar['date'], format='%Y-%m-%d')
calendar.head()
```

```python
#the column of price has value for the listings that are available and are null of the remain, 
#if we want to do the average of the prices for the all year is necessary to delete the null values

#delete the null values and the column 'available', that now has only 't' values
calendar_clean = calendar.dropna()
calendar_clean.pop('available')

#we need to conver the values on the 'price' column to number, to be able to do the average
calendar_price = calendar_clean.copy()
calendar_price['price'] = calendar_clean['price'].replace('[\$,]', '', regex=True).astype(float)
calendar_price.head()
```

```python
#we now want to the the average of the prices for each day
calendar_mean_price = calendar_price.groupby(['date'])['price'].mean()
calendar_mean_price.head()
```

```python
#create a plot to better observe the mean prices obtained
mean_prices = calendar_mean_price.plot()
#format the date axis to present only the month
date_form = DateFormatter("%m")
mean_prices.xaxis.set_major_formatter(date_form)

plt.show()
```

```python
#creating listings table
listings_short = listings[['id','accommodates','price']]
listings_short = listings_short.rename(columns={'id':'listing_id'})
listings_short = listings_short.copy()
listings_short['price'] = listings_short['price'].replace('[\$,]', '', regex=True).astype(float)

#creating calendar table
calendar = calendar.copy()
calendar['price'] = calendar['price'].replace('[\$,]', '', regex=True).astype(float)

#creating the merged table
merged_table = calendar.merge(listings_short, on='listing_id', how='left')
merged_table = merged_table.rename(columns={'price_x':'day_price','price_y':'average_price'})
merged_table
```

```python
merged_table['week'] = merged_table.apply(lambda x : x['date'].isocalendar()[1],axis=1)
merged_table
```

```python
merged_table.day_price.fillna(merged_table.average_price, inplace=True)
merged_table
```

```python
merged_table = merged_table.copy()
merged_table['price_accommodation'] = merged_table.day_price / merged_table.accommodates
merged_table
```

```python
#we now want to the the average of the prices for each day
calendar_mean_price_accommodation = merged_table.groupby(['date'])['price_accommodation'].mean()
calendar_mean_price_accommodation.head()

#create a plot to better observe the mean prices obtained
mean_prices_accommodation = calendar_mean_price_accommodation.plot()
#format the date axis to present only the month
date_form = DateFormatter("%m")
mean_prices_accommodation.xaxis.set_major_formatter(date_form)

plt.show()
```

```python
#we now want to the the average of the prices for each day
calendar_mean_week_price = merged_table.groupby(['week'])['day_price'].mean()
calendar_mean_week_price.head()

#create a plot to better observe the mean prices obtained
calendar_mean_week_price_plt = calendar_mean_week_price.plot()
#format the date axis to present only the month
plt.show()
```
