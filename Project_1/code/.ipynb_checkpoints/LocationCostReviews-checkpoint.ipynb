```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import seaborn as sns
import matplotlib.pyplot as plt
#is necessary to install the PyShp library: pip install pyshp
import shapefile as shp
```

```python
#load the listings data
listings = pd.read_csv('listings.csv')
listings.shape
```

```python
#reduce the data to just the columns of the cleaning fee and the cleaning score
lst_location_reviews = listings[['id','price','neighbourhood_cleansed','review_scores_rating','review_scores_location']].copy()
#droping the NaN values
lst_location_reviews = lst_location_reviews.dropna()
#convert the 'price' columns from string to number
lst_location_reviews = lst_location_reviews.copy()
lst_location_reviews['price'] = lst_location_reviews['price'].replace('[\$,]', '', regex=True).astype(float)
print(lst_location_reviews.head())
```

```python
#Aggregate the prices by prices and rates
lst_location_reviews_mean = lst_location_reviews.groupby(['neighbourhood_cleansed'])[['price','review_scores_rating','review_scores_location']].mean()
lst_location_reviews_mean
```

```python
lst_location_reviews_mean.to_csv(r'neighbourhood_price_reviews_average.csv')
```

```python
#reading the shape file by using reader function of the shape lib
sp = shp.Reader("Seattle_Neighborhoods.shp")
```

```python
def plot_shape(id, s=None):
    plt.figure()
    #plotting the graphical axes where map ploting will be done
    ax = plt.axes()
    ax.set_aspect('equal')
#storing the id number to be worked upon
    shape_ex = seattle_map.shape(id)
#NP.ZERO initializes an array of rows and column with 0 in place of each elements 
    #an array will be generated where number of rows will be(len(shape_ex,point))and number of columns will be 1 and stored into the variable
    x_lon = np.zeros((len(shape_ex.points),1))
#an array will be generated where number of rows will be(len(shape_ex,point))and number of columns will be 1 and stored into the variable
    y_lat = np.zeros((len(shape_ex.points),1))
    for ip in range(len(shape_ex.points)):
        x_lon[ip] = shape_ex.points[ip][0]
        y_lat[ip] = shape_ex.points[ip][1]
#plotting using the derived coordinated stored in array created by numpy
    plt.plot(x_lon,y_lat) 
    x0 = np.mean(x_lon)
    y0 = np.mean(y_lat)
    plt.text(x0, y0, s, fontsize=10)
# use bbox (bounding box) to set plot limits
    plt.xlim(shape_ex.bbox[0],shape_ex.bbox[2])
    return x0, y0
```

```python
def plot_map(sf, x_lim = None, y_lim = None, figsize = (10,14)):
    plt.figure(figsize = figsize)
    id=0
    for shape in sf.shapeRecords():
        x = [i[0] for i in shape.shape.points[:]]
        y = [i[1] for i in shape.shape.points[:]]
        plt.plot(x, y, 'k')
        
        if (x_lim == None) & (y_lim == None):
            x0 = np.mean(x)
            y0 = np.mean(y)
            plt.text(x0, y0, id, fontsize=10)
        id = id+1
    
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim)
        
    plt.grid()
#calling the function and passing required parameters to plot the full map
seattle_map = plot_map(sp)
```
